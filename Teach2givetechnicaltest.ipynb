  **QUESTION 2**

Write a Python function that checks whether a word or phrase is palindrome or not.  

Note: A palindrome is word, phrase, or sequence that reads the same  backward as forward, e.g., madam, kayak, racecar, or a phrase "nurses run"

```python
def check_is_palindrome(string):
  # Defines a function that rearranges the characters in a string from right to left and compares if the results are the same.
  left_to_right=[]

  # Breaks the string into individual characters
  for character in string:
    if character != ' ': # Accounts for whitespaces in strings
      left_to_right.append(character.lower()) # Accounts for uppercase characters
  print('letters from left to right:', left_to_right)

  # Reordering the characters from the last item in the list to the first
  right_to_left = left_to_right[::-1]
  print('letters from right to left:', right_to_left)

  # Checks if the two lists are the same
  return print('is palindrome:', left_to_right == right_to_left)
```

```python
# Example of a non-palindrome
check_is_palindrome('elephant')
```

```python
# Example of a palindrome word
check_is_palindrome('Kayak')
```

```python
# Example of a palindrome phrase
check_is_palindrome('nurses run')
```

  **QUESTION 3**

 Write a Python function to check whether a string is pangram or not. (Assume the string passed in does not have any punctuation)

 Note: Pangrams are words or sentences containing every letter of the  alphabet at least once. For example: "The quick brown fox jumps over the lazy dog"

```python
def check_is_pangram(sentence):
  # Defines a function that checks whether all the letters of the alphabet are in a sentence

  import string
  # The string module has the entire alphabet, negating the need to manually type out a list of 26 characters
  alphabet=list(string.ascii_lowercase)
  print('the alphabet:', alphabet)

  # Creating a list of all the characters in the sentence in lower case
  letters=[letter for letter in sentence.lower()]
  print('sentence:', letters)

  # Checking that all the letters in the alphabet are present in the sentence
  return print('is pangram:', all(character in letters for character in alphabet))
```

```python
# Example of a pangram
check_is_pangram('The quick brown fox jumps over the lazy dog')
```

```python
# Example of a pangram
check_is_pangram('Glibjocksquiznymphtovexdwarf')
```

```python
# Example of a non-pangram
check_is_pangram('I think, therefore I am')
```

  **QUESTION 4**
  

Write a program that takes an integer as input and returns an integer with reversed digit ordering.
Examples:
For input 500, the program should return 5.
For input -56, the program should return -65.
For input -90, the program should return -9.
For input 91, the program should return 19.

```python
def reverse_number(number):
  # Defines a function that reverses an integer

  # Accounting for negative integers
  if number < 0:
    # Coverting the integer into a string without the minus symbol
    string_number = str(abs(number))
    # Reverting the number back to negative after flipping it right to left
    reverse = (int((string_number[::-1])) * -1)

  # Accounting for positive integers
  else:
    string_number = str(number)
    reverse = int(string_number[::-1])
  return reverse
```

```python
# Example of a positive integer
reverse_number(91)
```

```python
# Example of a negative integer
reverse_number(-65)
```

```python
# Example of a zeroed number
reverse_number(500)
```

  **QUESTION 5**

Write a program that accepts a string as input, capitalizes the first letter of each word in the string, and then returns the result string.

Examples: "hi"=> returns "Hi"
"i love programming"=> returns "I Love Programming"

```python
def capitalize(sentence):
  '''
    The module string.capwords() splits the sentence into words using str.split(),
    capitalizes each word using str.capitalize(),
    & joins the capitalized words using str.join().

  '''
  import string
  return string.capwords(sentence)
```

```python
capitalize('i love programming')
```

```python
%%sql
SELECT
  cir.location,
  res.grid,
  res.position,
  res.fastestLap AS fastest_lap,
  res.points,
  (SELECT dri.forename || ' ' || dri.surname) AS driver_name,
  rac.name AS race_name,
  res.time,
  rac.year,
  con.name AS team_name,
  rac.date
FROM
  circuits AS cir
JOIN races AS rac ON cir.circuitid = rac.circuitid
JOIN results AS res ON rac.raceid = res.raceid
JOIN drivers AS dri ON res.driverid = dri.driverid
JOIN constructors AS con ON res.constructorid = con.constructorid
WHERE year = 2020
ORDER BY points DESC;
```
